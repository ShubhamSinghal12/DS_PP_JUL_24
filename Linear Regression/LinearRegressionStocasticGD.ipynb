```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_csv = pd.read_csv("./Training Data/Linear_X_Train.csv").values
y_csv = pd.read_csv('./Training Data/Linear_Y_Train.csv').values
# x_csv.head()
```

```python
plt.scatter(x_csv,y_csv)
plt.show()
```

```python
def hypothesis(x,th1,th0):
    return th1*x+th0

def error(x,th1,th0,y):
    y_pred = hypothesis(x,th1,th0)
    return np.mean((y_pred-y)**2)

def gradientDescent(x,y,lr=0.01,epochs = 1000):
    th1 = np.random.randn()
    th0 = np.random.randn()
    # print(th1,th0)
    errorGD = []
    for i in range(epochs):
        #Stochastic Gradient Descent
        j = np.random.randint(0,x.shape[0])
        y_pred = hypothesis(x[j],th1,th0)
        #Gradient
        
        th0 = th0 - lr*(y_pred-y[j])
        th1 = th1 - lr*((y_pred-y[j])*x[j])

        errorGD.append(error(x,th1,th0,y))
        print(f"Epoch: {i+1}, Error: {errorGD[-1]}, r2 score{r2_score(y,y_pred)}")
        
        # print(th1,th0)

    return th1,th0,errorGD

def r2_score(y_true,y_pred):
    rss = np.sum((y_true-y_pred)**2)
    tss = np.sum((y_true-np.mean(y_true))**2)
    return 1 - rss/tss

```

```python
th1,th0,errorGD = gradientDescent(x_csv,y_csv)
```
