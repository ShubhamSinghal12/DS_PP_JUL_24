```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,0.1)
```

```python
def fn(x):
    return 2*x**2 - 3*x + 4

def detF(x):
    return 4*x-3
```

```python
y = fn(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
xm = 8
lr = 0.4
pltXM = []
epochs = 100
for i in range(epochs):
    pltXM.append(xm)
    xm = xm - lr*detF(xm)
xm
```

```python
plt.plot(x,y)
plt.plot(pltXM,fn(np.array(pltXM)))
plt.show()
```
