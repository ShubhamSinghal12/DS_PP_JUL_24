```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_csv = pd.read_csv("./Training Data/Linear_X_Train.csv").values
y_csv = pd.read_csv('./Training Data/Linear_Y_Train.csv').values
# x_csv.head()
```

```python
plt.scatter(x_csv,y_csv)
plt.show()
```

```python
def hypothesis(x,th1,th0):
    return th1*x+th0

def error(x,th1,th0,y):
    y_pred = hypothesis(x,th1,th0)
    return np.mean((y_pred-y)**2)

def gradientDescent(x,y,lr=0.01,epochs = 1000):
    th1 = np.random.randn()
    th0 = np.random.randn()
    # print(th1,th0)
    for i in range(epochs):
        y_pred = hypothesis(x,th1,th0)
        # print(y_pred)
        #Gradient
        th0 = th0 - lr*np.mean(y_pred-y)
        th1 = th1 - lr*np.mean((y_pred-y)*x)
        # print(th1,th0)

    return th1,th0
```

```python
th1,th0 = gradientDescent(x_csv,y_csv)
```
