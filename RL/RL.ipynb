```python
import gymnasium as gym
```

```python
env = gym.make('CartPole-v1',render_mode = 'human')
print(env.reset())
for i in range(100):
    env.render()
    action = env.action_space.sample()  # Take a random action
    obs, reward, done, truncated, info = env.step(action)
    print(obs)
    if done or truncated:
        env.reset()
        print()
env.close()
```
