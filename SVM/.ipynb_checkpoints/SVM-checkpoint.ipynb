```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,random_state=3)
```

```python
y[y == 0] = -1
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
class SVM:
    def __init__(self,C=0.01,lr= 0.01,epochs = 1000):
        self.C = C
        self.lr = lr
        self.epochs = epochs

    def fit(self,X,y):
        self.th,self.b,self.errorGD = self.gradient_descent(X,y)

    def predict(self,X):
        y_pred = self.th@X.T+self.b
        y_pred[y_pred >= 0] = 1
        y_pred[y_pred < 0] = -1
        return y_pred
    def hinge_loss(self,X,y,th,b):
        ti = self.functional_margin(X,y,th,b)
        er = 1-ti
        er[er < 0] = 0
        return 0.5*(th@th) + self.C*np.sum(er)

    def functional_margin(self,X,y,th,b):
        return y*(th@X.T+b)
    
    def gradient_descent(self,X,y):

        th = np.zeros(X.shape[1])
        b = 0
        errorGD = []
        for i in range(self.epochs):
            errorGD.append(self.hinge_loss(X,y,th,b))
            
            ti = self.functional_margin(X,y,th,b)
            ti[ti < 1] = -1
            ti[ti >= 1] = 0
            gradient = th + self.C*((ti*y)@X)

            th = th - self.lr*gradient
            b = b - self.lr*(self.C*np.sum(ti*y))

            if i%100 == 0:
                print(f"Epoch: {i+1}, Error: {errorGD[-1]}")

        return th,b,errorGD
     
    def accuracy_score(self,y_true,y_pred):
        return (y_true==y_pred).mean()
        
```

```python
clf = SVM(C = 0.01,epochs = 201)
clf.fit(X,y)
clf.accuracy_score(y,clf.predict(X))
```

```python
plt.plot(clf.errorGD)
plt.show()
```

```python
x1 = np.linspace(-1,3,100)
x2 = -(clf.th[0]*x1+clf.b)/clf.th[1]

plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(x1,x2,c='b')
plt.plot(x1,x2+1/np.linalg.norm(clf.th,ord=2),c='r')
plt.plot(x1,x2-1/np.linalg.norm(clf.th,ord=2),c='r')
plt.show()
```
