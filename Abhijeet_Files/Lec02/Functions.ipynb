```python
def reportcard(marks):
# marks = 82

    if marks > 90:
        return ('A+')
    elif marks > 80:
        return ('A')
    elif marks > 70:
        return ('B+')
    elif marks > 60:
        return ('B')
    elif marks > 50:
        return ('C+')
    elif marks > 40:
        return ('C')
    else:
        return ('F')        
```

```python
print(reportcard(52))
```

```python
def print_largest(var1, var2, var3):
    if (var1 > var2 and var1 > var3):
        print(var1)
    elif (var2 > var1 and var2 > var3):
        print(var2)
    else:
        print(var3)
```

```python
print_largest(10, 20, 65)
```
